```python
import pandas as pd

df = pd.read_csv('Depression.csv')
```

```python
print(df['Depression Label'].unique())
```

```python
# List of values to encode as 1
positive_values = ['Moderate Depression','Severe Depression', 'Moderately Severe Depression']

# Encode: 1 if in list, else 0
df['Depression Label'] = df['Depression Label'].isin(positive_values).astype(int)
```

```python
X = df.drop(['1. Age','2. Gender','3. University','4. Department','5. Academic Year', '6. Current CGPA','7. Did you receive a waiver or scholarship at your university?','Depression Value','Depression Label'], axis=1)
y = df['Depression Label']
```

```python
X
```

```python
# Custom list of new names (must match number of columns)
new_names = [i for i in range(9)]
X.columns = new_names
```

```python
# Compute cumulative sums across question columns row-wise
numeric_cols = [i for i in range(9)]
cumsum_df = X[numeric_cols].cumsum(axis=1)

# Rename cumulative sum columns as S1, S2, ..., S5
cumsum_df.columns = [f'S{i+1}' for i in range(len(cumsum_df.columns))]

cumsum_df
```

```python
cumsum_df.to_csv('output.csv', index=False)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

def find_best_thresholds(cumsum_df, y):
    thresholds = {}

    for col in cumsum_df.columns:
        best_acc = 0
        best_thresh = None

        # Try all unique values as candidate thresholds
        for t in range(0,30):
            preds = (cumsum_df[col] >= t).astype(int)
            acc = accuracy_score(y, preds)
            if acc > best_acc:
                best_acc = acc
                best_thresh = t

        thresholds[col] = {'threshold': best_thresh, 'accuracy': best_acc}

    return thresholds

# ✅ Assuming cumsum_df contains S1 to S5 and y is the label Series
thresholds = find_best_thresholds(cumsum_df, y)

# ✅ Print thresholds and their accuracy
for col, result in thresholds.items():
    print(f"{col}: threshold = {result['threshold']}, accuracy = {result['accuracy']:.2f}")
```

```python
import pandas as pd
from sklearn.metrics import roc_curve

# Assume df is your DataFrame, and 'label' is the binary target column
label_col = 'Depression Label'

# Automatically get all numeric columns except the label
feature_cols = cumsum_df.select_dtypes(include='number').columns

# Store thresholds
threshold_results = {}

# Loop through each feature column
for col in feature_cols:
    y_true = df[label_col]
    y_scores = cumsum_df[col]

    fpr, tpr, thresholds = roc_curve(y_true, y_scores)

    # Youden’s J statistic
    optimal_idx = (tpr - fpr).argmax()
    optimal_threshold = thresholds[optimal_idx]

    threshold_results[col] = optimal_threshold
    print(f"{col}: Best threshold = {optimal_threshold:.4f}")

# Convert to DataFrame (optional)
threshold_df = pd.DataFrame.from_dict(threshold_results, orient='index', columns=['Best Threshold'])
```
