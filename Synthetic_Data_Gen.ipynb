```python
import pandas as pd
import random

N = 1000

def generate_scores_anxiety_depression(length, level):
    scores = []
    
    for _ in range(length):
        if level == 0:  
            value = random.choices([0, 1])[0]
        
        elif level == 1:
            value = random.choices([1, 2, 3])[0]
        
        else:           
            value = random.choices([2, 3])[0]
        
        if random.random() < 0.1: 
            if level == 0:
                value = random.choice([1, 2])
            elif level == 1:
                value = random.choice([0, 1])
            else:
                value = random.choice([0, 1])
        
        scores.append(value)

    return scores


def generate_scores_stress(length, level):
    scores = []
    
    for _ in range(length):
        if level == 0: 
            value = random.choices([0, 1, 2])[0]
        
        elif level == 1:
            value = random.choices([1, 2, 3])[0]
        
        else:         
            value = random.choices([3, 4])[0]
        
        if random.random() < 0.1:
            if level == 0:
                value = random.choice([3, 4])
            elif level == 1:
                value = random.choice([0, 4])
            else:
                value = random.choice([0, 1])
        
        scores.append(value)

    return scores

anxiety_data = []
depression_data = []
stress_data = []

for _ in range(N):
    anxiety_level = random.choice([0, 1, 2])
    depression_level = random.choice([0, 1, 2])
    stress_level = random.choice([0, 1, 2])

    anxiety = generate_scores_anxiety_depression(7, anxiety_level)
    depression = generate_scores_anxiety_depression(9, depression_level)
    stress = generate_scores_stress(10, stress_level)

    anxiety_data.append(anxiety)
    depression_data.append(depression)
    stress_data.append(stress)

anxiety_df = pd.DataFrame(anxiety_data, columns=[f"Anxiety_{i+1}" for i in range(7)])
depression_df = pd.DataFrame(depression_data, columns=[f"Depression_{i+1}" for i in range(9)])
stress_df = pd.DataFrame(stress_data, columns=[f"Stress_{i+1}" for i in range(10)])

anxiety_df.to_csv("anxiety_subset.csv", index=False)
depression_df.to_csv("depression_subset.csv", index=False)
stress_df.to_csv("stress_subset.csv", index=False)

print("Data saved to CSV files.")
```
